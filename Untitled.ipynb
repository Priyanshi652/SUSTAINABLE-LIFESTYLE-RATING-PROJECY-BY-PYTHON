1.Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

2.Reading the data

```python
d_f=pd.read_csv("lifestyle_sustainability_data.csv")
d_f
```

```python
d_f.shape
```

```python
d_f.info()
```

3.Data Cleaning

```python
d_f.isnull().sum()
```

```python
#filling the null values-:
for i in d_f.select_dtypes(include="object").columns:
    d_f[i].fillna(d_f[i].mode()[0],inplace=True)
```

```python
d_f.isnull().sum()
```

Droping unwanted columns from dataset


```python
d_f.drop(["ParticipantID"],axis=1,inplace=True)
```

```python
d_f.drop(["Rating"],axis=1,inplace=True)
```

```python
d_f.columns
```

```python
d_f.describe()
```

4.Data Visualization alonh with insights

```python
#1.CALCULATING AVERAGE AGE of people following diet-type
average_age=d_f.groupby('DietType')['Age'].mean()
average_age.round(1)
sns.barplot(x=average_age.index,y=average_age.values,palette="dark",width=0.5)
plt.title("AVERAGE AGE OF PEOPLE BY DIET TYPE")
plt.ylabel("Average ages")
plt.show()
```

```python
#2.PEOPLE OF WHICH LOCATION USES PUBLIC TRANSPORT MOdE
public_transport=d_f[d_f["TransportationMode"]=="Public Transit"]
location_percentage=public_transport["Location"].value_counts()
plt.pie(location_percentage,labels=location_percentage.index,autopct="%1.1f%%")
plt.title("PERCENTAGE OF PUBLIC TRANSIT BY DIFFERENT LOCATION")
plt.show()
```

```python
#3.local food frequency in urban and suburban
location_data=d_f[d_f["Location"].isin(["Urban","Rural"])]
sns.countplot(x="Location",hue="UsingPlasticProducts",data=location_data,palette="magma",width=0.8)
plt.title("USE OF PLASTIC PRODUCTS BY URBAN AND RURAL AREA")
plt.show()
```

```python
#how much clothing frequency have suistainable brand-:
pivot_table=d_f.pivot_table(index="ClothingFrequency",columns="SustainableBrands",aggfunc="size",fill_value=0)
sns.heatmap(pivot_table,annot=True,cmap="YlGnBu",fmt="d")
plt.title("SUSTAINALBLE BRANDS BY CLOTHING FREQUENCY")
plt.show()
```

```python
#age of people involved in community on basis of gender
community
orders=["Low","Moderate","High"]
sns.barplot(data=d_f,x="CommunityInvolvement",y="Age",hue="Gender",palette="viridis",order=orders)
plt.title("COMMMUNITY INVOLVEMENT BY AGE")
plt.show()
```

```python
# average electricity is consumed by different home type
group=d_f.groupby('HomeType')['MonthlyElectricityConsumption'].mean()
group.round(1)
```

```python
result=pd.DataFrame({'MonthlyElectricityConsumption':group})
plt.plot(result,c="b",marker="*",markersize=9,linewidth=2,mfc="green",mec="green")
plt.xlabel("HOME TYPES",c="brown",size=10)
plt.ylabel("ELECTICTY CONSUMED",c="brown",size=10)
plt.title("ELECTRICITY CONSUMED BY HOME TYPES")
plt.grid()
```
